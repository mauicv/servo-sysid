```python
import json
import os

path = os.getcwd().split('/')[:-1]
filename = '/'.join(path) + '/dataset/actions-dataset.json'

with open(filename, 'r') as f:
    data = json.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np

actions = np.array(data['data'][0]['actions'])

plt.plot(actions)
```

```python
count = 0
for item in data['data']:
    if item['type'] == 'step':
        count += 1
        actions = np.array(item['actions'])
        plt.plot(actions)


plt.show()
assert count == 10, f"count is {count} but should be 8"
```

```python
fig, axs = plt.subplots(ncols=3, nrows=1)
count = 0
for item in data['data']:
    if item['type'] == 'prbs':
        actions = np.array(item['actions'])
        axs[count].plot(actions)
        count += 1

assert count == 3, f"count is {count} but should be 3"
```
